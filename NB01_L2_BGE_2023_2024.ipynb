## Quelques rappels sur les variables
Les variables en Python sont des « valeurs modifiables stockées dans la mémoire de l'ordinateur », comme les informations de vos contacts dans la mémoire de votre téléphone.

Elles peuvent être simples ou complexes :
- quelques exemples de types simples : 
    - booléens (`True` ou `False`) valeurs de vérité vrai ou faux ;
    - nombres entiers (dans trois représentations différentes, nous n'utiliserons que la représentation décimale classique) 
    | Base                   | Représentation | Valeur    |
    | ---------------------- | -------------- | --------- |
    | décimale (base 10)     | `15`           | 15        |
    | octale (base 8)        | `0o15`         | 1x8+5=13  |
    | hexadécimale (base 16) | `0x15`         | 1x16+5=21 |
    - nombres réels : ils peuvent être écrits sous plusieurs formes simples : 
      - `-21.1` (attention au point à la place de la virgule)
      - `43.25e+3` vaut 43250.0 
      - `43.25e+17` vaut 4.325e+18
      - `32.3e-4` et `32.3E-4` valent 0.00323
    - nombres complexes : ils comprennent une partie rélle et une partie imaginaire, par exemple `4.3 + 2.7j` (attention au `j` à la place du `i` français de l'unité imaginaire), `-2.3j` pour une partie réelle nulle, `3.1 + 0j` pour une partie imaginaire nulle. 
    - chaînes de caractères ;
- quelques exemples de types complexes : 
    - listes (ensemble de valeurs modifiable, les valeurs peuvent être hétérogènes) ; 
    - dictionnaires (ensemble de couples (clé,valeur) nous en reparlerons) 
    - tuples (ensemble de valeurs non modifiable), informations relatives à un contact (prénom, nom, téléphone).

Comme je l'ai indiqué dans le cours, les variables sont utilisées par le programme pour faire des calculs par exemple.

Lorsqu'en mathématiques on définit la fonction $f$ par : 
$f(x)=3.2x-2cos(x)+1.3$ on trouve dans cette définition : 
- des valeurs constantes : les réels $3.2$ et $1.3$, l'entier $2$ ;
- des opérations implicites ($3.2x$ correspond à $3.2\times x$) ou non (la soustraction $3.2x-2cos(x)$) ;
- l'appel à une fonction prédéfinie : $cos(x)$ ;
- une variable muette $x$ à laquelle on fait correspondre son image par la fonction $f$.

### Variables : affichage du type et de l'emplacement de stockage mémoire
Le type de la variable dépend de la valeur rentrée : dans l'exemple précédent, `2` est compris par Python comme étant la valeur entière $2$ alors que `2.` serait considérée comme la valeur réelle $2,0$

À chaque variable est associée un emplacement de stockage en mémoire : la quantité de mémoire totale utilisée par un programme Python dépend du nombre de variables utilisées et de leurs types

````python
# Affichage du type de la variable i
print(f"Type de la variable i : {type(i)}")

# Affichage de l'emplacement de stockage en mémoire 
print(f"Emplacement de stockage en mémoire : {id(i)}")
````

### Variables : affichage du contenu d'une variable

Le but de ce que nous faisons en programmation, c'est d'effectuer des calculs mais évidemment il faut, à un moment ou à un autre, afficher le résultat des calculs effectués par le programme.

Il existe plusieurs façons d'afficher le contenu des variables (leurs valeurs).

On utilise, dans tous les cas présentés ici, la fonction `print` avec en *argument* la chaîne de caractères que l'on souhaite afficher.
- ligne 3 : chaîne de caractères suivie d'une virgule et du nom de la variable dont on souhaite afficher le contenu.
- **ligne 5 : `f` précise le format d'affichage complet (avec entre `{}` le nom de la variable dont on souhaite afficher le contenu à la place de son nom). [Documentation officielle de Python vers cette forme que nous utiliserons le plus souvent possible](https://docs.python.org/fr/3.10/tutorial/inputoutput.html#formatted-string-literals)**
- ligne 6 : chaîne d'affichage complet avec indication de la forme de l'affichage `{:d}` signifie au format « entier » suivie du mot clé Python `format` et du nom de la variable dont on souhaite afficher le contenu.
- ligne 7 : chaîne d'affichage complet avec indication de la forme de l'affichage `{:6d}` signifie au format « entier sur 6 chiffres, compléter à gauche par des espaces si besoin » suivie du mot clé Python `format` et du nom de la variable dont on souhaite afficher le contenu.

````python
# Plusieurs façons d'afficher la valeur de la variable i

print("Valeur de i :",i)
# La forme que nous utiliserons le plus souvent possible : la f-string
print(f"Valeur de i : {i}")
print("Valeur de i : {:d}".format(i))
print("Valeur de i : {:6d}".format(i))

print(f"{i} est plus petit que 10 : {i<10}")  
````

### Variables : lecture du contenu au clavier et stockage dans une variable

Pour la lecture d'une valeur au clavier, on utilise la fonction prédéfinie `input` qui reçoit entre parenthèses la chaîne de caractères à afficher à l'écran pour inviter l'utilisateur à saisir un texte ou une valeur numérique. Cette fonction *retourne* la valeur saisie au clavier par l'utilisateur (ici la valeur sera transformée en nombre réel avec le **`float`** ).


````python
x=float(input("Entrez une valeur réelle :"))
# Entrez au clavier la valeur 12.2 par exemple

print(f"Le carré de {x} vaut {x*x}")
````

### Comment faire des calculs ?
Nous avons écrit que les variables étaient utilisées pour faire des calculs, comme par exemple lorsqu'à partir du lycée, on vous a expliqué que pour résoudre l'équation :

$ax^2+bx+c=0$

il fallait calculer un $\Delta=b^2-4ac$ puis en fonction du signe de $\Delta$ etc

Dans cet exemple, on utilise des variables ($a$, $b$, $c$) qui seront utilisées pour déterminer $\Delta$ et qui finalement, grâce à un **algorithme** permettront de déterminer le nombre de racines/solutions possibles et leurs valeurs tel que résumé ci-dessous.

|  Discriminant | Nombre et nature des racines |
| :------------- | :--------------------------------------- |
| $\Delta >0$   | 2 racines réelles : $x_1=\frac{-b-\sqrt{\Delta}}{2a}$ et $x_2=\frac{-b+\sqrt{\Delta}}{2a}$ |
| $\Delta=0$    | 1 racine réelle (double) : $x_1=x_2=\frac{-b}{2a}$ |
| $\Delta <0$   | 2 racines complexes conjuguées : $x_1=\frac{-b-i\sqrt{\Delta}}{2a}$ et $x_2=\frac{-b+i\sqrt{\Delta}}{2a}$ |


[Lien vers les éléments du cours sur les opérateurs arithmétiques si besoin](https://codimd.univ-rouen.fr/s/STHMOTTdi#Quelques-op%C3%A9rateurs-arithm%C3%A9tiques-de-base)


````python
# Affichage à l'écran de la phrase "Entrer x :" 
# stockage dans la variable x de la valeur (réelle=float) lue au clavier
x=float(input("Entrer x :"))

# Affichage à l'écran de la phrase "Entrer n :" 
# stockage dans la variable n de la valeur (entier=int) lue au clavier
n=int(input("Entrer n :"))

# Affichage de la puissance n-ième de x
print(f"{x}^{n}={x**n}")

# En Python, pour obtenir la racine carrée de x, on utilise np.sqrt(x) 
# sqrt est la fonction square root (racine carrée) du module prédéfini numpy (numeric python)
# Il convient d'importer le module numpy avant d'utiliser la fonction sqrt
import numpy as np

# Exemple pour la racine carrée de 2 (~ 1.4142)
x=2.
print(f"{x} a pour racine carrée {np.sqrt(x)}")
````

### Première utilisation informelle des listes

Par exemple : 


````python
# Exemples de liste :

notes=[8,12,16,11,15,20,6,14,11,18,9,15,17]

print(notes[1])
````

Qu'observez-vous lorsque vous exécutez la cellule précédente ?

`notes[1]` vaut par conséquent 12 et non 8 comme on aurait pu croire.
