## Quelques rappels sur les variables
Les variables en Python sont des  valeurs modifiables stockes dans la mmoire de l'ordinateur , comme les informations de vos contacts dans la mmoire de votre tlphone.

Elles peuvent tre simples ou complexes :
- quelques exemples de types simples : 
    - boolens (`True` ou `False`) valeurs de vrit vrai ou faux ;
    - nombres entiers (dans trois reprsentations diffrentes, nous n'utiliserons que la reprsentation dcimale classique) 
    | Base                   | Reprsentation | Valeur    |
    | ---------------------- | -------------- | --------- |
    | dcimale (base 10)     | `15`           | 15        |
    | octale (base 8)        | `0o15`         | 1x8+5=13  |
    | hexadcimale (base 16) | `0x15`         | 1x16+5=21 |
    - nombres rels : ils peuvent tre crits sous plusieurs formes simples : 
      - `-21.1` (attention au point  la place de la virgule)
      - `43.25e+3` vaut 43250.0 
      - `43.25e+17` vaut 4.325e+18
      - `32.3e-4` et `32.3E-4` valent 0.00323
    - nombres complexes : ils comprennent une partie rlle et une partie imaginaire, par exemple `4.3 + 2.7j` (attention au `j`  la place du `i` franais de l'unit imaginaire), `-2.3j` pour une partie relle nulle, `3.1 + 0j` pour une partie imaginaire nulle. 
    - chanes de caractres ;
- quelques exemples de types complexes : 
    - listes (ensemble de valeurs modifiable, les valeurs peuvent tre htrognes) ; 
    - dictionnaires (ensemble de couples (cl,valeur) nous en reparlerons) 
    - tuples (ensemble de valeurs non modifiable), informations relatives  un contact (prnom, nom, tlphone).

Comme je l'ai indiqu dans le cours, les variables sont utilises par le programme pour faire des calculs par exemple.

Lorsqu'en mathmatiques on dfinit la fonction $f$ par : 
$f(x)=3.2x-2cos(x)+1.3$ on trouve dans cette dfinition : 
- des valeurs constantes : les rels $3.2$ et $1.3$, l'entier $2$ ;
- des oprations implicites ($3.2x$ correspond  $3.2\times x$) ou non (la soustraction $3.2x-2cos(x)$) ;
- l'appel  une fonction prdfinie : $cos(x)$ ;
- une variable muette $x$  laquelle on fait correspondre son image par la fonction $f$.

### Variables : affichage du contenu d'une variable

Le but de ce que nous faisons en programmation, c'est d'effectuer des calculs mais videmment il faut,  un moment ou  un autre, afficher le rsultat des calculs effectus par le programme.

Il existe plusieurs faons d'afficher le contenu des variables (leurs valeurs).

On utilise, dans tous les cas prsents ici, la fonction `print` avec en *argument* la chane de caractres que l'on souhaite afficher.
- ligne 3 : chane de caractres suivie d'une virgule et du nom de la variable dont on souhaite afficher le contenu.
- **ligne 5 : `f` prcise le format d'affichage complet (avec entre `{}` le nom de la variable dont on souhaite afficher le contenu  la place de son nom). [Documentation officielle de Python vers cette forme que nous utiliserons le plus souvent possible](https://docs.python.org/fr/3.10/tutorial/inputoutput.html#formatted-string-literals)**
- ligne 6 : chane d'affichage complet avec indication de la forme de l'affichage `{:d}` signifie au format  entier  suivie du mot cl Python `format` et du nom de la variable dont on souhaite afficher le contenu.
- ligne 7 : chane d'affichage complet avec indication de la forme de l'affichage `{:6d}` signifie au format  entier sur 6 chiffres, complter  gauche par des espaces si besoin  suivie du mot cl Python `format` et du nom de la variable dont on souhaite afficher le contenu.

````python
# Plusieurs faons d'afficher la valeur de la variable i

print("Valeur de i :",i)
# La forme que nous utiliserons le plus souvent possible : la f-string
print(f"Valeur de i : {i}")
print("Valeur de i : {:d}".format(i))
print("Valeur de i : {:6d}".format(i))

print(f"{i} est plus petit que 10 : {i<10}")  
````

### Comment faire des calculs ?
Nous avons crit que les variables taient utilises pour faire des calculs, comme par exemple lorsqu' partir du lyce, on vous a expliqu que pour rsoudre l'quation :

$ax^2+bx+c=0$

il fallait calculer un $\Delta=b^2-4ac$ puis en fonction du signe de $\Delta$ etc

Dans cet exemple, on utilise des variables ($a$, $b$, $c$) qui seront utilises pour dterminer $\Delta$ et qui finalement, grce  un **algorithme** permettront de dterminer le nombre de racines/solutions possibles et leurs valeurs tel que rsum ci-dessous.

|  Discriminant | Nombre et nature des racines |
| :------------- | :--------------------------------------- |
| $\Delta >0$   | 2 racines relles : $x_1=\frac{-b-\sqrt{\Delta}}{2a}$ et $x_2=\frac{-b+\sqrt{\Delta}}{2a}$ |
| $\Delta=0$    | 1 racine relle (double) : $x_1=x_2=\frac{-b}{2a}$ |
| $\Delta <0$   | 2 racines complexes conjugues : $x_1=\frac{-b-i\sqrt{\Delta}}{2a}$ et $x_2=\frac{-b+i\sqrt{\Delta}}{2a}$ |


[Lien vers les lments du cours sur les oprateurs arithmtiques si besoin](https://codimd.univ-rouen.fr/s/STHMOTTdi#Quelques-op%C3%A9rateurs-arithm%C3%A9tiques-de-base)


````python
# Affichage  l'cran de la phrase "Entrer x :" 
# stockage dans la variable x de la valeur (relle=float) lue au clavier
x=float(input("Entrer x :"))

# Affichage  l'cran de la phrase "Entrer n :" 
# stockage dans la variable n de la valeur (entier=int) lue au clavier
n=int(input("Entrer n :"))

# Affichage de la puissance n-ime de x
print(f"{x}^{n}={x**n}")

# En Python, pour obtenir la racine carre de x, on utilise np.sqrt(x) 
# sqrt est la fonction square root (racine carre) du module prdfini numpy (numeric python)
# Il convient d'importer le module numpy avant d'utiliser la fonction sqrt
import numpy as np

# Exemple pour la racine carre de 2 (~ 1.4142)
x=2.
print(f"{x} a pour racine carre {np.sqrt(x)}")
````
